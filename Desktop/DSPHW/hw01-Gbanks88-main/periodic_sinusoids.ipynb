Determine whether each of the following signals is periodic. If the signal is periodic, state its period.

Respond with variables `periodic = True/False` and `N_0 = XX`.  If not periodic set `N_0 = 0`.

1. $x[n] = e^{j(\pi n/6)}$

```python
import numpy as np
import matplotlib.pyplot as plt

n_vals_2 = np.arange(-12, 13)
x_2 = np.exp(1j * (np.pi * n_vals_2 / 6))
periodic_2 = True  # This signal is periodic
N_0_2 = 12  # Period is 12

plt.subplot(3, 1, 2)
plt.stem(n_vals_2, np.real(x_2), linefmt='b-', markerfmt='bo', basefmt='r-')
plt.stem(n_vals_2, np.imag(x_2), linefmt='g-', markerfmt='go', basefmt='r-')
plt.title('Signal 2: $x[n] = e^{j(\pi n / 6)}$')
plt.legend(['Real Part', 'Imaginary Part'])
plt.grid(True)
```

```python
assert isinstance(periodic, bool)
assert isinstance(N_0, int)
```

2. $x[n] = e^{j(3πn/4)}$

```python
assert isinstance(periodic, bool)
assert isinstance(N_0, int)
```

3. $x[n] = \frac{\sin(\pi n/5)}{\pi n}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of n
n_vals = np.arange(-20, 21)  # Range from -20 to 20

# Compute the signal
x = np.sinc(n_vals / 5)  # sinc function handles sin(πx)/(πx)

# Plot the signal
plt.stem(n_vals, x, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title('$x[n] = \\frac{\\sin(\\pi n / 5)}{\\pi n}$')
plt.grid(True)
plt.show()
```
