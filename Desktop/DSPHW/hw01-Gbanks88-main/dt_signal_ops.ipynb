# Discrete-Time Signal Operations

```python
from matplotlib import pyplot as plt
import numpy as np
```

A discrete-time signal $x[n]$ is shown below:

<img src="dt_signal_ops_fig.png" width="350"/>

Create an array for the following signal values from $n = -2,\dots,6$ for the shifted signal $\tilde{x}[n]$. E.g., to represent the figure above, we would use `x = np.array([0,1,1,1,1,1,0.5,0,0])`.

1. $\tilde{x}[n]=x[n âˆ’ 2]$

```python
import numpy as np
import matplotlib.pyplot as plt

# Original signal x[n] for n = -2 to 6
x = np.array([0, 1, 1, 1, 1, 1, 0.5, 0, 0]).astype(float)

# Define the range of n for the shifted signal
n_vals = np.arange(-2, 7)

# Create the shifted signal \tilde{x}[n] = x[n - 2]
x_tilde = np.array([x[n - 2] if (n - 2) >= -2 and (n - 2) <= 6 else 0 for n in n_vals])

# Check the length of x_tilde
assert len(x_tilde) == 9

# Plot the original and shifted signals
plt.figure()
plt.stem(n_vals, x, label="$x[n]$")
plt.stem(n_vals, x_tilde, markerfmt="rx", linefmt="r-", label="$\\tilde{x}[n] = x[n - 2]$")
plt.legend()
plt.xlabel("$n$")
plt.ylabel("$x[n]$")
plt.title("Original and Shifted Signals")
plt.grid(True)
plt.show()
```

```python
assert len(x_tilde) == 9
n_vals = np.arange(-2, 7)
x = np.array([0, 1, 1, 1, 1, 1, 0.5, 0, 0]).astype(float)

plt.figure()
plt.stem(n_vals, x, label="$x[n]$")
plt.stem(n_vals, x_tilde, markerfmt="rx", linefmt="r-", label="$\\widetilde{x}[n]$")
plt.legend()
plt.xlabel("$n$")

```

5. $\tilde{x}[n]=x[n-1]\delta [n-3]$

```python
# Dirac delta function Î´[n-3]
delta = np.array([1 if n == 3 else 0 for n in n_vals])

# Signal \tilde{x}[n] = x[n-1]\delta[n-3]
x_tilde_3 = np.array([x[n-1] if (n-1) >= -2 and (n-1) <= 6 else 0 for n in n_vals]) * delta

# Check length
assert len(x_tilde_3) == 9

# Plot
plt.figure()
plt.stem(n_vals, x, label="$x[n]$")
plt.stem(n_vals, x_tilde_3, markerfmt="rx", linefmt="r-", label="$\\tilde{x}[n] = x[n-1]\delta[n-3]$")
plt.legend()
plt.xlabel("$n$")
plt.ylabel("$x[n]$")
plt.title("Original and Delta-Multiplied Signals")
plt.grid(True)
plt.show()
```
